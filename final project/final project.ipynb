we see there are plently of outliers based on the IQR, so we would need to remove it

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df_new = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_new
```

```python
df_new = df.copy()
for col in ['PRICE', 'BEDS', 'BATH', 'PROPERTYSQFT']:
    df_new = remove_outliers(df_new, col)
```
